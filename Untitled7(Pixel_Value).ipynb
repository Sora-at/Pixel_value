```python
import rasterio
from rasterio import plot
import matplotlib
import matplotlib.pyplot as plt
import folium
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
ndvi = rasterio.open('D:/test1/NDVI2/NDVI_test4.tif')
```

```python
plot.show(ndvi)
```

```python
ndvi.read(1).astype(float) #define data type = .astype(float)
```

```python
np.min(ndvi.read(1)), np.max(ndvi.read(1)) #find value min,max of array
```

```python
fig, ax = plt.subplots(figsize=(12,8))
axisx_labels = [i for i in range(1,30,1)] 
axisy_labels = [i for i in range(1,30,1)]
sns.heatmap(ndvi.read(1),linewidths=0,annot=True,cmap='YlGn',fmt='.2f',annot_kws={"size": 7}) 


ax.set_xticklabels(axisx_labels, rotation=90, ha='center', fontsize=8)
ax.set_yticklabels(axisy_labels, rotation=0, fontsize=8)
ax.xaxis.tick_top();

```

```python
#Delete row 
#a=np.delete(ndvi.read(1),0,0)
#Delete rows and columns
#(np.delete(np.delete(a, 1, 0), 1, 1)) 
a=(np.delete(np.delete(ndvi.read(1), 0, 1), 0, 0)) 
```

```python
fig, ax = plt.subplots(figsize=(12,8))
axisx_labels = [i for i in range(700635,700636)]
axisy_labels = [i for i in range(1686888,1686889)]

sns.heatmap(a,linewidths=0,annot=True,cmap='YlGn',fmt='.2f',annot_kws={"size": 8})

ax.set_xticklabels(axisx_labels, rotation=0, ha='center', fontsize=8)
ax.set_yticklabels(axisy_labels, rotation=0, fontsize=8)
ax.xaxis.tick_top();
ax.set_title("Example NDVI")
```
